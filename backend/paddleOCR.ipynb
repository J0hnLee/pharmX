```python
from paddleocr import PaddleOCR, draw_ocr

# 同样也是通过修改 lang 参数切换语种
ocr = PaddleOCR(lang="chinese_cht",use_gpu=False) # 首次执行会自动下载模型文件
img_path = './forTest/202109.png'
#img_path=otsu_th
result = ocr.ocr(img_path)
# 可通过参数控制单独执行识别、检测
result = ocr.ocr(img_path, det=True) #只执行识别
#result = ocr.ocr(img_path, rec=True) #只执行检测
# 打印检测框和识别结果
for line in result:
    print(line)

# 可视化
from PIL import Image
image = Image.open('./forTest/202109.png').convert('RGB')
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]
im_show = draw_ocr(image, boxes, txts, scores,font_path='./forTest/chinese_cht.ttf')
im_show = Image.fromarray(im_show)
im_show.save('./forTest/result.jpg')
```

```python
import cv2
import numpy as np
import sys
from matplotlib import pyplot as plt
```

```python
img_path = './forTest/S__25207736.jpg'

img=cv2.imread(img_path,0)
#ret2, otsu_th = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

th1=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,7,2)
images=[img,th1]
plt.figure()
for i in range(2):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
plt.show
cv2.imwrite('./forTest/output.jpg', th1)
```
